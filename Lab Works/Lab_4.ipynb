```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
  
#loading image
img_lena = cv2.imread('Lenna_(test_image).png')
#Converting RBG to Grayscale and saving as Gslenna.jpg
gray_img = cv2.cvtColor(img_lena, cv2.COLOR_BGR2GRAY)
cv2.imwrite('Lenna_(test_image).jpg',gray_img)
plt.imshow(gray_img,cmap='gray')
plt.title('GsLenna')
```

```python
dst = cv2.calcHist(gray_img, [0], None, [256], [0,256])

plt.hist(gray_img.ravel(),256,[0,256])
plt.title('Histogram for gray scale image')
plt.show()
```

```python
img_lena.min()*255
img_lena.max()*255
ratio=((255-0)/(141-52))+0
print(ratio)
LC=(img_lena-np.min(img_lena))*ratio

dst = cv2.calcHist(img_lena, [0], None, [256], [0,256])

plt.hist(LC.ravel(),256,[0,256])
plt.title('Histogram for an linear corrected image')
plt.show()
```

```python
def gammaCorrection(src, gamma):
    invGamma = 1 / gamma
 
    table = [((i / 255) ** invGamma) * 255 for i in range(256)]
    table = np.array(table, np.uint8)

    return cv2.LUT(src, table)
 
img = cv2.imread('Lenna_(test_image).jpg')
gammaImg = gammaCorrection(img, 0.5)

f=plt.figure(figsize=(15,15))
f.add_subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
f.add_subplot(1,2,2)
plt.imshow(gammaImg)
plt.title('Gamma corrected image')

```
