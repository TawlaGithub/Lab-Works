```python
image=cv2.imread('images.jpg')
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(image.shape)
```

```python
plt.imshow(gray,cmap='gray')
plt.title('Royal family of BHUTAN')
```

```python
# counting the number of pixels
number_of_white_pix = np.sum(image == 255)
print(number_of_white_pix)
```

```python
Higest_pix= np.amax(image)
print(Higest_pix)
```

```python
#covert image to black from white
thresh,image_black=cv2.threshold(gray,255,0,cv2.THRESH_BINARY)
cv2.imshow('images.jpg',image_black)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title('Royal family of BHUTAN')
```

```python
cv2.rectangle(image,(50,50),(200,200),(0,255,0),3)
roi=image(50:200,50:200)
cv2.imshow("resized image",roi)
roi_pro=roi.shape,roi.size
print(roi_pro)
file=open('desktop.text','W')
file.write('resolution:(150,150,3)\nSize:67500')
file.close()
```

```python
blue=image[:,:,0]
green=image[:,:,1]
red=image[:,:,2]
cv2.imshow('Blue pixels',blue)
cv2.imshow('Green pixels',green)
cv2.imshow('Red pixels',red)
cv2.imwrite('images.jpg',image_black)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b,g,r=cv2.split(image)
cv2.imshow('Blue pixels',b)
cv2.imshow('Green pixels',g)
cv2.imshow('Red pixels',r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_merged=cv2.merge(b,g,r)
cv2.imshow('Blue pixels',image_merged)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
